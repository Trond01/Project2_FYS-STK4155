```python
import sys, os
sys.path.append(os.path.abspath(os.path.join('..')))

# Import tools from this project
from Code.activation_functions import sigmoid, relu, leaky_relu

import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.25
x = np.linspace(-7, 7, 100)

fig, ax = plt.subplots(1, 3, figsize=(9, 3))

ax[0].plot(x, sigmoid(x))
ax[1].plot(x, relu(x))
ax[2].plot(x, leaky_relu(x, alpha))

for i, tit in enumerate(["Sigmoid", "RELU", fr"Leaky RELU, $\alpha={alpha}$"]):
    ax[i].set(title=tit)

plt.tight_layout()
plt.show()

plt.savefig("Figures/activation_functions.png")
```
