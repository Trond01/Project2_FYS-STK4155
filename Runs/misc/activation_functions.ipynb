```python
import sys, os
sys.path.append(os.path.abspath(os.path.join('..\..')))

# Import tools from this project
from Code.activation_functions import sigmoid, relu, leaky_relu

import matplotlib.pyplot as plt
import numpy as np

from jax import numpy as jnp
import seaborn as sns
sns.set()

import matplotlib as mpl
mpl.rcParams['font.size'] = 16
mpl.rcParams["axes.labelsize"] = 16
mpl.rcParams["axes.titlesize"] = 18
mpl.rcParams['legend.fontsize'] = 14
mpl.rcParams['font.family'] = 'sans-serif'
plt.rcParams['xtick.labelsize']=14
plt.rcParams['ytick.labelsize']=14
```

```python
alpha = 0.25
x = np.linspace(-7, 7, 100)

fig, ax = plt.subplots(1, 4, figsize=(12, 3))

ax[0].plot(x, sigmoid(x))
ax[1].plot(x, relu(x))
ax[2].plot(x, leaky_relu(x, alpha))
ax[3].plot(x, jnp.tanh(x))

for i, tit in enumerate(["Sigmoid", "ReLu", fr"Leaky ReLu, $\alpha={alpha}$", "tanh"]):
    ax[i].set(title=tit)

plt.tight_layout()
plt.savefig("../Figures/activation_functions.png")
plt.show()
```
